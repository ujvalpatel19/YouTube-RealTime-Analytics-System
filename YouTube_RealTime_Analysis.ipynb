# 📊 YouTube Real-Time Analytics System

## 📦 Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 📂 Step 2: Load Dataset

```python
df = pd.read_csv("youtube_data.csv")
df.head()
```

## ℹ️ Step 3: Dataset Information

```python
print('Shape:', df.shape)
print(df.columns)
df.info()
```

## 🧹 Step 4: Missing Values Check

```python
df.isnull().sum()
```

## 🔥 Step 5: Top Channels Analysis
Identify most active channels in trending videos.

```python
df['channel'].value_counts().head(10)
```

## 👀 Step 6: Top Viewed Videos
Find most popular videos based on views.

```python
df.sort_values(by='views', ascending=False).head(10)
```

## 💯 Step 7: Engagement Ratio
Measure performance using likes + comments vs views.

```python
df['engagement'] = (df['likes'] + df['comments']) / df['views']
df[['title','engagement']].sort_values(by='engagement', ascending=False).head()
```

## 📊 Step 8: Views Distribution

```python
plt.figure()
df['views'].hist()
plt.title('Views Distribution')
plt.show()
```

## 📈 Step 9: Views vs Likes Relationship

```python
plt.figure()
plt.scatter(df['views'], df['likes'])
plt.xlabel('Views')
plt.ylabel('Likes')
plt.title('Views vs Likes')
plt.show()
```

## 🔗 Step 10: Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=['int64','float64'])
plt.figure()
sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
